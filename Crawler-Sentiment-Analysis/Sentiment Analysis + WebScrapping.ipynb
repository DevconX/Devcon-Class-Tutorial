```python
train = imdb[(imdb['type'] ==  'train') & (imdb['label'] != 'unsup')]
```

```python
from sklearn import metrics
```

```python
from sklearn.feature_extraction import stop_words
 
print(stop_words.ENGLISH_STOP_WORDS)
```

```python
cv = CountVectorizer(stop_words = 'english') 
```

```python
bow = cv.fit(train['review'])
```

```python
bow_transform =bow.transform(train['review'])
```

```python
classifier_mnb = MultinomialNB().fit(bow_transform, train['label'])
```

```python
test = imdb[(imdb['type'] == 'test') & (imdb['label'] != 'unsup')]
```

```python
bow_comments = bow.transform(test['review'])
```

```python
from sklearn import metrics
```

```python
predictions = classifier_mnb.predict(bow_comments)
```

```python
metrics.accuracy_score(test['label'], predictions)
```

```python
from urllib.request import Request, urlopen as uReq
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
def make_soup(website):
    
    req = Request(website, headers= {'User-Agent' : 'Mozilla/5.0'}) 
    uClient = uReq(req)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, "html.parser")
    return page_soup
```

```python
GOOGLE_NEWS_URL = 'https://www.google.com.my/search?q={}&source=lnt&tbm=nws'

def forge_url(q):
    return GOOGLE_NEWS_URL.format(q.replace(' ','+'))
```

```python
news_url = forge_url('avengers')
```

```python
website = make_soup(news_url)
```

```python
website.h3
```

```python
website.findAll('h3')
```

```python
website.h3.a
```

```python
website.h3.a.text
```

```python
news_headlines = pd.DataFrame(columns = ['title', 'sentiment', 'score'])
```

```python
news_headlines
```

```python
headlines = website.findAll("h3")
```

```python

n = 0
for item in headlines: 
    top = item.a
    text_headlines = top.text
    print(text_headlines)
    n += 1
    news_headlines.loc[n-1, 'title'] = text_headlines
```

```python
news_headlines.head()
```

```python
news_headlines.head()
```

```python
n = 0 
for i in news_headlines['title']:
    print(i)
    n += 1
    bow_comments = bow.transform([i])
    predictions = classifier_mnb.predict(bow_comments)
    score = round(classifier_mnb.predict_proba(bow_comments)[0][1],5)
    news_headlines.loc[n-1,'sentiment'] = predictions
    news_headlines.loc[n-1, 'score'] = score
    print(predictions)
    print(score)
    
```
